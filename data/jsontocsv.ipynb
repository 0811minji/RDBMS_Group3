```python
import os
import json
import pandas as pd

path = '/Users/kimseonhyeong/Desktop/SKKU/2024년 2학기/관계형데이터베이스/Project/Use/[원천]성취수준데이터셋_train'

out_path = '/Users/kimseonhyeong/Desktop/SKKU/2024년 2학기/관계형데이터베이스/Project/data'
```

```python
answerCode_list = []
test_IRT_list = []
learner_IRT_list = []

for grade_folder in os.listdir(path):

    if grade_folder != '.DS_Store':

        grade_folder_path = os.path.join(path, grade_folder)

        for exam_folder in os.listdir(grade_folder_path):

            if exam_folder != '.DS_Store':

                exam_folder_path = os.path.join(grade_folder_path, exam_folder)
                answerCode_folder_path = os.path.join(exam_folder_path, '1_문항정오답표')
                test_IRT_folder_path = os.path.join(exam_folder_path, '2_문항IRT')
                learner_IRT_folder_path = os.path.join(exam_folder_path, '3_응시자IRT')

                
                for file in os.listdir(answerCode_folder_path):
                    
                    file_path = os.path.join(answerCode_folder_path, file)

                    with open(file_path, 'r') as f:
                        js = json.load(f)

                    df = pd.json_normalize(js)
                    answerCode_list.append(df)
                
                for file in os.listdir(test_IRT_folder_path):
                    
                    file_path = os.path.join(test_IRT_folder_path, file)

                    with open(file_path, 'r') as f:
                        js = json.load(f)

                    df = pd.json_normalize(js)
                    test_IRT_list.append(df)

                for file in os.listdir(learner_IRT_folder_path):
                    
                    file_path = os.path.join(learner_IRT_folder_path, file)

                    with open(file_path, 'r') as f:
                        js = json.load(f)

                    df = pd.json_normalize(js)
                    learner_IRT_list.append(df)

answerCode_df = pd.concat(answerCode_list, ignore_index=True)
test_IRT_df = pd.concat(test_IRT_list, ignore_index=True)
learner_IRT_df = pd.concat(learner_IRT_list, ignore_index=True)
```

```python
learner_df = answerCode_df[['learnerID', 'learnerProfile']]
learner_df.sort_values('learnerID', inplace=True, ignore_index=True)
learner_df
```

```python
answerCode_df.drop('learnerProfile', axis=1, inplace=True)
answerCode_df.sort_values(['learnerID', 'testID', 'assessmentItemID'], inplace=True, ignore_index=True)
answerCode_df
```

```python
test_IRT_df.sort_values(['testID', 'assessmentItemID'], inplace=True, ignore_index=True)
test_IRT_df
```

```python
learner_IRT_df.drop('learnerProfile', axis=1, inplace=True)
learner_IRT_df.sort_values(['learnerID', 'testID'], inplace=True, ignore_index=True)
learner_IRT_df
```

```python
learner_df.to_csv(os.path.join(out_path, 'learner.csv'), index=False)
answerCode_df.to_csv(os.path.join(out_path, 'answerCode.csv'), index=False)
test_IRT_df.to_csv(os.path.join(out_path, 'test_IRT.csv'), index=False)
learner_IRT_df.to_csv(os.path.join(out_path, 'learner_IRT.csv'), index=False)
```
