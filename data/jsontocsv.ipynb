```python
import os
import json
import pandas as pd

path = './Use/[원천]성취수준데이터셋_train'

out_path = './Data'
```

```python
answerCode_list = []
IRT_list = []

for grade_folder in os.listdir(path):

    if grade_folder != '.DS_Store':

        grade_folder_path = os.path.join(path, grade_folder)

        for exam_folder in os.listdir(grade_folder_path):

            if exam_folder != '.DS_Store':

                exam_folder_path = os.path.join(grade_folder_path, exam_folder)
                answerCode_folder_path = os.path.join(exam_folder_path, '1_문항정오답표')
                IRT_folder_path = os.path.join(exam_folder_path, '2_문항IRT')
                
                for file in os.listdir(answerCode_folder_path):
                    
                    file_path = os.path.join(answerCode_folder_path, file)

                    with open(file_path, 'r') as f:
                        js = json.load(f)

                    df = pd.json_normalize(js)
                    answerCode_list.append(df)
                
                for file in os.listdir(IRT_folder_path):
                    
                    file_path = os.path.join(IRT_folder_path, file)

                    with open(file_path, 'r') as f:
                        js = json.load(f)

                    df = pd.json_normalize(js)
                    IRT_list.append(df)

answerCode_df = pd.concat(answerCode_list, ignore_index=True)
IRT_df = pd.concat(IRT_list, ignore_index=True)
```

```python
answerCode_df.sort_values(['learnerID', 'testID', 'assessmentItemID'], inplace=True, ignore_index=True)
answerCode_df
```

```python
IRT_df.sort_values(['testID', 'assessmentItemID'], inplace=True, ignore_index=True)
IRT_df
```

```python
knowledge_path = './Use/[라벨]수학 지식체계 데이터 세트_210611.json'

with open(knowledge_path, 'r') as f:
    js = json.load(f)

first_level = {key: js[key] for key in js}

knowledge_list = []

for key, item in first_level.items():
    
    fromConcept_df = pd.json_normalize(item['fromConcept'])
    toConcept_df = pd.json_normalize(item['toConcept'])
    toConcept_df.rename(columns={'id':'toConceptid'}, inplace=True)

    df = pd.concat([fromConcept_df, toConcept_df['toConceptid']], axis=1)
    knowledge_list.append(df)

knowledge_df = pd.concat(knowledge_list, ignore_index=True)

knowledge_df
```

```python
answerCode_df.to_csv(os.path.join(out_path, 'answerCode.csv'), index=False)
IRT_df.to_csv(os.path.join(out_path, 'IRT.csv'), index=False)
knowledge_df.to_csv(os.path.join(out_path, 'knowledge.csv'), index=False)
```
