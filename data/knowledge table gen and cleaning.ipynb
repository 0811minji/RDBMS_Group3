```python
import os
import json
import pandas as pd

out_path = './data'
```

```python
knowledge_path = '/Users/minjiku/Desktop/RDBMS/수학분야 학습자 역량 측정 데이터/수학지식체계 데이터셋/[라벨]수학 지식체계 데이터 세트_210611.json'

with open(knowledge_path, 'r') as f:
    js = json.load(f)

first_level = {key: js[key] for key in js}

knowledge_list = []

for key, item in first_level.items():
    
    fromConcept_df = pd.json_normalize(item['fromConcept'])
    toConcept_df = pd.json_normalize(item['toConcept'])
    prereqConcept = toConcept_df.rename(columns={'id':'prereqConcept.id'})

    from_df = pd.concat([fromConcept_df, prereqConcept['prereqConcept.id']], axis=1)
    knowledge_list.append(from_df)
    
    if ~toConcept_df['id'].isin(from_df['id']).any():
        to_df = pd.concat([toConcept_df, None], axis=1)
        knowledge_list.append(to_df)

df = pd.concat(knowledge_list, ignore_index=True)
df
```

```python
import numpy as np

# 'semester' column을 '학습학년', '학습학기'로 분리하기
df['course_grade'] = np.nan
df['course_semester'] = np.nan

def process_semester(row):
        parts = row.split('-')
        
        # 고등학교 특수 케이스
        if parts[0] == '고등':
            if parts[1] == '수학':
                if parts[2] == '상':
                    return 10, 1
                elif parts[2] == '하':
                    return 10, 2
            elif parts[1] == '수1':
                return 11, 1
            elif parts[1] == '수2':
                return 11, 2
            else:  # 기하, 미적, 확통
                return np.nan, np.nan
        
        # 초등학교, 중학교 케이스
        else:
            grade = parts[1][1]  # 초1, 중2 등에서 숫자만 추출
            if parts[0] == '중등': 
                grade = int(grade)+6
            
            semester = 1 if '1학기' in parts[2] else 2 if '2학기' in parts[2] else np.nan
            return grade, semester
    
# 데이터 처리 적용
df[['course_grade', 'course_semester']] = pd.DataFrame(
    df['semester'].apply(process_semester).tolist(),
    index=df.index)

df
```

```python
subsequent_concepts = df.groupby('prereqConcept.id')['id'].agg(list).to_dict()

df['subsequentConcept.id'] = df['id'].map(lambda x: subsequent_concepts.get(x, np.nan))

df
```

```python
prerequisite_concepts = df.groupby('id')['prereqConcept.id'].agg(list).to_dict()

df['prereqConcept.id'] = df['id'].map(lambda x: prerequisite_concepts.get(x, np.nan))

df
```

```python
df.columns
```

```python
df.drop_duplicates(subset=['id', 'name', 'semester', 'description', 'chapter.id', 'chapter.name',
       'achievement.id', 'achievement.name','course_grade', 'course_semester'], inplace=True)
```

```python
df['chapter_lv1'] = np.nan
df['chapter_lv2'] = np.nan
df['chapter_lv3'] = np.nan

df[['chapter_lv1', 'chapter_lv2', 'chapter_lv3']] = df['chapter.name'].str.split('>', expand=True).apply(lambda x: x.str.strip())

df
```

```python
# 리스트 안의 nan 값을 제거
df['prereqConcept.id'] = df['prereqConcept.id'].apply(lambda x: [item for item in x if pd.notna(item)] if isinstance(x, list) else x)
```
