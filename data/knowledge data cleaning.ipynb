```python
import pandas as pd
import numpy as np

df = pd.read_csv("knowledge.csv")

#선행개념 칼럼 이름 다시 짓기
df.rename(columns={'toConceptid':'prereqConcept.id'}, inplace=True)

df['semester'].unique()
```

```python
# 'semester' column을 '학습학년', '학습학기'로 분리하기
df['course_grade'] = np.nan
df['course_semester'] = np.nan

def process_semester(row):
        parts = row.split('-')
        
        # 고등학교 특수 케이스
        if parts[0] == '고등':
            if parts[1] == '수학':
                if parts[2] == '상':
                    return 10, 1
                elif parts[2] == '하':
                    return 10, 2
            elif parts[1] == '수1':
                return 11, 1
            elif parts[1] == '수2':
                return 11, 2
            else:  # 기하, 미적, 확통
                return np.nan, np.nan
        
        # 초등학교, 중학교 케이스
        else:
            grade = parts[1][1]  # 초1, 중2 등에서 숫자만 추출
            if parts[0] == '중등': 
                grade = int(grade)+6
            
            semester = 1 if '1학기' in parts[2] else 2 if '2학기' in parts[2] else np.nan
            return grade, semester
    
# 데이터 처리 적용
df[['course_grade', 'course_semester']] = pd.DataFrame(
    df['semester'].apply(process_semester).tolist(),
    index=df.index)

df
```

```python
subsequent_concepts = df.groupby('prereqConcept.id')['id'].agg(list).to_dict()

df['subsequentConcept_id'] = df['id'].map(lambda x: subsequent_concepts.get(x, np.nan))

df
```

```python
df.to_csv('knowledge.csv', index=False)
```
