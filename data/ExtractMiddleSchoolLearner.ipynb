```python
import os
import pandas as pd

path = '/Users/kimseonhyeong/Desktop/SKKU/2024년 2학기/관계형데이터베이스/Project/data'
```

```python
learner = pd.read_csv(os.path.join(path, 'learner.csv'))

learner = learner[learner['learnerID'].str[2].astype(int)>=7].reset_index(drop=True)

learner[['gender', 'unused', 'grade']] = learner['learnerProfile'].str.split(";", expand=True)

learner['gender'] = learner['gender'].map({'F':0, 'M':1})
learner['grade'] = learner['grade'].astype(int)

learner.drop(['learnerProfile', 'unused'], axis=1, inplace=True)
learner
```

```python
answerCode = pd.read_csv(os.path.join(path, 'answerCode.csv'))
answerCode = answerCode[answerCode['learnerID'].str[2].astype(int)>=7].reset_index(drop=True)
answerCode
```

```python
learner_IRT = pd.read_csv(os.path.join(path, 'learner_IRT.csv'))
learner_IRT = learner_IRT[learner_IRT['learnerID'].str[2].astype(int)>=7].reset_index(drop=True)
learner_IRT
```

```python
learner.to_csv(os.path.join(path, 'ms_learner.csv'), index=False)
answerCode.to_csv(os.path.join(path, 'ms_answerCode.csv'), index=False)
learner_IRT.to_csv(os.path.join(path, 'ms_learner_IRT.csv'), index=False)
```
